```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('Area(sq.ft)')
plt.ylabel('Prices(USA)')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']].values,df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('Area(sq.ft)')
plt.ylabel('Prices(USA)')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```

```python
area = pd.read_csv("area.csv")
area
```

```python
price = reg.predict(area)
```

```python
area['Price'] = price
```

```python
area
```

```python
area.to_csv("Prediction.csv",index=False)
```
